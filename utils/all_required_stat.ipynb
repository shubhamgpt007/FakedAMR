### 3—TTR

```python
# Function to calculate the type-to-token ratio of a tweet
def ttr(tweet):
    tokens = nltk.word_tokenize(tweet.lower())
    types = set(tokens)
    return len(types) / len(tokens)

# Apply the ttr function to the 'Tweet' column and create a new 'TTR' column
df = df.assign(TTR=df['tweet'].apply(ttr))
```

### 24—Longest sequence of consecutive vowels

```python
# Function to find longest sequence of consecutive vowels in a string
def longest_vowel_sequence(text):
    pattern = re.compile(r'[aeiouAEIOU]+')
    matches = pattern.findall(text)
    if len(matches) == 0:
        return 0
    else:
        return max(len(match) for match in matches)

# Apply the longest_vowel_sequence function to the 'Tweet' column and create a new 'Longest Vowel Sequence' column
df['longest_vowel_sequence'] = df['tweet'].apply(lambda x: longest_vowel_sequence(x))
```

```python
# List of determiners to count
determiners = ['a', 'an', 'the', 'this', 'that', 'these', 'those', 'my', 'your', 'his', 'her', 'its', 'our', 'their', 'few', 'many', 'much', 'most', 'several', 'some', 'any', 'enough', 'all', 'both', 'half', 'either', 'neither', 'each', 'every', 'other', 'another', 'such']

# Function to count determiners in a tweet
def count_determiners(tweet):
    doc = nlp(tweet)
    determiner_count = 0
    for token in doc:
        if token.pos_ == 'DET' and token.text.lower() in determiners:
            determiner_count += 1
    return determiner_count

# Apply the count_determiners function to the 'Tweet' column and create a new 'Number of Determiners' column
df['number_of_determiners'] = df['tweet'].apply(lambda x: count_determiners(x))
```

### 37—Tweet entropy

```python
from scipy.stats import entropy
import string
```

```python
# Function to calculate tweet entropy
def tweet_entropy(tweet):
    # Convert tweet to lowercase
    tweet = tweet.lower()
    # Remove punctuation
    tweet = tweet.translate(str.maketrans('', '', string.punctuation))
    # Calculate frequency of each character
    freq = [tweet.count(c) for c in set(tweet)]
    # Calculate probability distribution
    prob = [f/sum(freq) for f in freq]
    # Calculate entropy
    ent = entropy(prob)
    return ent

# Apply the tweet_entropy function to the 'Tweet' column and create a new 'Tweet Entropy' column
df['tweet_entropy'] = df['tweet'].apply(lambda x: tweet_entropy(x))
```
